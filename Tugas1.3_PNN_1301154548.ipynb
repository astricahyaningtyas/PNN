SUMMATION LAYER
========

```python
def summation_layer(test, label):
    labels = np.unique(label)
    sum_ = np.zeros((test.shape[0], labels.shape[0]))
    for i, _ in enumerate(test):
        g_ = pattern_layer(_)
        for j in labels:
            sum_[i, int(j)] = np.sum(g_[label[:]==j])
    return sum_

_summation = summation_layer(test, dlabel_train)
_summation
```

OUTPUT LAYER
======

```python
def output_layer(summation):
    return np.array([np.argmax(_) for _ in summation])

label  = np.array(output_layer(_summation))
label
```
